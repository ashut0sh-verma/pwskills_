```python
#Q14

date_range = pd.date_range(start='2023-01-01', periods=10)
temperature = np.random.uniform(low=15, high=30, size=len(date_range))
humidity = np.random.uniform(low=30, high=90, size=len(date_range))
df = pd.DataFrame({ 'Date': date_range,'Temperature': temperature,'Humidity': humidity})
df

```

```python

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(df.index, df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='red')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()
ax2.plot(df.index, df['Humidity'], color='green', label='Humidity')
ax2.set_ylabel('Humidity', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
#Q15
np.random.seed(42)
data = np.random.normal(scale =1 ,loc=0, size = 1000)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, bins =30, color='orange' ,stat='density', edgecolor='black' , kde=False)
mean = np.mean(data)
std = np.std(data)
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

plt.plot(x, pdf, color='red', linewidth=2, label='Normal Distribution PDF')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Q17
n=10
x = np.random.uniform(-10, 10, n)
y = np.random.uniform(-10, 10, n)

df = pd.DataFrame({'X': x, 'Y': y})
```

```python
df
```

```python
def get_quadrant(row):
    if row['X'] >= 5 and row['Y'] >= 5:
        return 'Quadrant 1'
    elif row['X'] < 5 and row['Y'] >= 5:
        return 'Quadrant 2'
    elif row['X'] < 5 and row['Y'] < 5:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

df['Quadrant'] = df.apply(get_quadrant, axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', s=100)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant wise Scatter Plot')
plt.legend(title='Quadrants')
plt.grid(True , alpha=0.3)
plt.show()
```

```python
#Q19

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=len(categories))
df = pd.DataFrame({'Category': categories, 'Value': values})
df
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral11
source = ColumnDataSource(df)
fig = figure(x_range=df['Category'], title='Random Categorical Bar Chart', x_axis_label='Category', y_axis_label='Value', width=800, height=400)

bars = fig.vbar(x='Category', top='Value', source=source, width=0.9, line_color='white', fill_color=factor_cmap('Category', palette=Spectral11, factors=categories))
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
fig.add_tools(hover)
show(fig)
```
