```python
#Q3
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 35, 20]
bars = plt.bar(categories , values , color = 'orange' , edgecolor = 'black' , width = 0.5 )

plt.grid(True ,axis='y' , alpha = 0.3)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f'{yval}', ha='center', va='bottom', fontsize=8)

plt.show()
```

```python
#Q4
data = np.random.normal(0, 1, 1000)
counts , bins, patches = plt.hist(data , bins = 30 , color = 'yellow' , edgecolor = 'black')
plt.grid(True , alpha = 0.3)
plt.title('Histogram')


for count, x in zip(counts, bins):
    plt.text(x + (bins[1] - bins[0]) / 2, count, str(int(count)),
             ha='center', va='bottom' , fontsize = 7)


plt.tight_layout()
plt.show()
```

```python
#Q3
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
#Q4
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python
df
```

```python
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),  x=df['X'].values.reshape(100, 100), y=df['Y'].values.reshape(100, 100))])
```

```python
#Q5
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
data = {'x': xx.flatten(), 'y': yy.flatten(), 'value': data_heatmap.flatten()}
df = pd.DataFrame(data)
```

```python
df
```

```python
mapper = LinearColorMapper(palette="Viridis256", low=df['value'].min(), high=df['value'].max())

fig = figure(title='Heatmap', width=1000, height=600)
fig.rect(x='x', y='y', width=0.12, height=0.13, source=df, fill_color=linear_cmap('value', palette="Viridis256", low=df['value'].min(), high=df['value'].max()), line_color=None)
color_bar = ColorBar(color_mapper=mapper, location=(0,0))
fig.add_layout(color_bar, 'right')
show(fig)
```
